**Enfermedad de Parkinson**

La enfermedad de Parkinson corresponde al desorden del movimiento más común y a la segunda enfermedad neurodegenerativa más frecuente del Sistema Nervioso Central (SNC). Predomina en hombres y la edad usual de presentación comprende entre los 65 y 70 años. En la neuropatología intervienen los niveles depletados de dopamina, secundario a las inclusiones citoplasmáticas de alfa-sinucleína. Los síntomas motores cardinales corresponden a la bradiquinesia, tremor en reposo y rigidez. Múltiples síntomas no motores se han descrito, los más reconocidos en el pródromo son la anosmia, la constipación y disturbios del sueño. El diagnóstico se realiza de forma clínica, la “International Parkinson and Movement Disorder Society” (MDS), proporciona criterios que pueden guiar al mismo. El principal biomarcador identificado corresponde a la alfa-Sinucleína. La terapia farmacológica está basada en preparaciones de Levodopa, agonistas dopaminérgicos e inhibidores de la monoamino oxidasa B. 

*Diagnóstico:*

El diagnóstico de la EP inicialmente está basado en la historia clínica y el examen físico. La historia puede proporcionar datos de antecedentes familiares de primer grado y síntomas motores o no motores. La “International Parkinson and Movement Disorder Society” (MDS), establece criterios designados a la investigación, pero pueden ser utilizados como guía diagnóstica, denominados “Clinical Diagnostic Criteria for Parkinson´s disease (MDS-PD Criteria)”. El
prerrequisito para aplicar los criterios corresponde al parkinsonismo, definido
como bradiquinesia en combinación con tremor en reposo, rigidez o ambos.
Para establecer la EP basados en la presentación clínica, los individuos requieren dos de cuatro criterios de soporte:
1. Tremor en reposo
2. Mejoría dramática con terapia dopaminérgica
3. Presencia de disquinesias inducidas por levodopa
4. Presencia de pérdida olfatoria o denervación simpática cardíaca
Los clínicos pueden considerar “banderas rojas” a aquellas manifestaciones no esperadas en la EP y que sugieran un diagnóstico alternativo al síndrome parkinsoniano. Algunas de estas manifestaciones incluyen, progresión rápida;
caídas, demencia, alucinaciones, fallas autonómicas y disfunción bulbar tempranas, oftalmoplejía vertical supranuclear y especialmente la ausencia robusta de respuesta a levodopa.
La patogénesis de la EP inicia años antes de la manifestación de los síntomas motores y del diagnóstico. Los biomarcadores podría utilizarse de maneras diagnóstica, predictiva y progresiva; podrían describir la severidad
de la enfermedad y colaborar en la elección del tratamiento. Idealmente, un
biomarcador óptimo debe poder cuantificarse en estadios tempranos de la
enfermedad, diferenciarse de forma efectiva de desordenes similares, ser costo efectivo y reproducible.
En investigaciones genómicas se asocia a la EP con variaciones en el gen SNCA, que codifica la alfa-Sinucleína, principal componente de los cuerpos de Lewy.
Al tener acceso al espacio extracelular, la alfa-sinucleína puede ser detectada en el líquido cefalorraquídeo (LCR), su nivel total está descendido en sinucleinopatías pero no es un hallazgo específico del Parkinson. La

*Reto del Diagnóstico Temprano:*

* Síntomas Iniciales: Los síntomas iniciales pueden ser sutiles y se confunden fácilmente con el envejecimiento normal o con otros trastornos.

**Enfermedad de Huntington**

La enfermedad de Huntington es un padecimiento neurológico degenerativo, de herencia autosómica dominante, causado por una expansión CAG que codifica una secuencia de poliglutamina en la proteína huntingtina. Su frecuencia varía de cinco a 10 afectados por 100 mil individuos en población caucásica. Clínicamente muestra manifestaciones motoras, cognoscitivas, psicológicas y muerte en 10 a 15 años. Avances concretos se han logrado en el conocimiento del mecanismo mutacional, alteraciones del producto proteico y su efecto neuropatológico. Un conjunto de procedimientos como PCR con o sin modificación del ADN, Southern blot y métodos mixtos son analizados en sus características y eficiencia para el diagnóstico molecular de esta enfermedad.

*Diagnóstico:*

La EH se puede identificar clínicamente en presencia de: alteraciones del comportamiento, alteraciones afectivas y cognitivas asociadas a una  disfunción motora progresiva y antecedentes familiares compatibles con una transmisión autosómica dominante.

La edad media de inicio de los síntomas es 38 años, con unos límites que pueden variar entre la segunda y la séptima décadas de la vida, la edad de inicio y gravedad de la enfermedad dependen  del tamaño de la expansión de la secuencia CAG 9,21-22.
 
De forma característica los signos iniciales son la inquietud generalizada, alteraciones del sueño, cambios en el comportamiento, así como ansiedad y depresión.

Trastornos  motores: al inicio los movimientos anormales son muy sutiles,  integrados en el seno de los  movimientos voluntarios o aparecen espontáneamente sin relación con estos. A medida que avanza la enfermedad los trastornos motores se hacen más prominentes, no son suprimibles y causan trastorno funcional. Con la evolución de la enfermedad  la corea empeora afectándose el equilibrio de la marcha y los movimientos voluntarios. En estadios avanzados puede aparecer parkinsonismo, a menudo por el tratamiento anticoreico, así como posturas distónicas. El lenguaje se torna disártrico por las hipercinesias, siendo la comprensión normal. La hiperreflexia es común y en el 10% de los casos hay signo de Babinski.

**Esclerosis lateral Amiotrófica**

La esclerosis lateral amiotrófica es un trastorno progresivo del sistema nervioso que gradualmente destruye los nervios responsables del movimiento muscular provocando una parálisis total de los músculos, incluyendo los de la respiración.

*Diagnóstico:*

La esclerosis lateral amiotrófica es un diagnóstico de exclusión. No hay ninguna prueba que pueda dar un diagnóstico definitivo de ELA, aunque la presencia de señales de deterioro de las neuronas motoras superiores e inferiores en una sola extremidad constituye un fuerte indicio. El diagnóstico de ELA está basado primordialmente en
los síntomas y señales obtenidos en la historia clínica y en el examen neurológico, el cual debe realizarse a intervalos regulares para evaluar si los síntomas están
empeorando progresivamente. Ya que los síntomas de la ELA pueden ser similares a aquellos de una gran variedad de otras enfermedades, hay que realizar los exámenes apropiados para descartar la posibilidad de otras enfermedades. Una de estas pruebas es la electromiografía para detectar la actividad eléctrica en los músculos y cuyos hallazgos
de la EMG pueden apoyar el diagnóstico de ELA. Otra prueba común mide la velocidad de conducción de los nervios cuyas anormalidades específicas pueden sugerir, por ejemplo, que el paciente tiene un tipo de neuropatía periférica o miopatía en vez de ELA. La Resonancia Magnética a menudo es normal en pacientes con ELA, pero es posible que revele problemas que puedan estar causando los síntomas, como un tumor en la médula espinal, una hernia discal cervical o espondilosis cervical. Otros diagnósticos diferenciales que deben excluirse son: Esclerosis Múltiple, tumor cerebral, deficiencia de vitamina B12, neuropatía motora multifocal, síndromes paraneoplásicos,intoxicación por plomo y polio.

*Reto del Diagnóstico Temprano:*

* Variabilidad de Síntomas: Los síntomas pueden variar ampliamente entre los pacientes y progresar a diferentes velocidades, lo que complica el diagnóstico.
* Falta de Biomarcadores Específicos: No hay pruebas de laboratorio específicas para confirmar el diagnóstico de ELA en sus primeras etapas.

**•** Discutir matemáticamente, explicando con fórmulas, cómo funciona la rutina detrend de scipy **(5%)** 


La rutina detrend de scipy se utiliza para eliminar tendencias lineales o polinómicas de las señales, lo que es especialmente útil en el análisis de señales fisiológicas. La función se basa en la idea de que las señales pueden contener una tendencia que puede enmascarar las variaciones de interés.

**Formula general**
Para una señal 𝑥(𝑡), la tendencia se puede expresar como:
<div style="text-align: center;">
    <h5>𝑥(𝑡)=𝑦(𝑡)+𝑓(𝑡)</h5>
</div>

donde:

• 𝑥(𝑡) es la señal observada.

• 𝑦(𝑡) es la señal de interés (sin tendencia).

• 𝑓(𝑡) es la tendencia (una función que varía en el tiempo).

El objetivo de la rutina *detrend* es encontrar y eliminar 𝑓(𝑡).

**Método de Detrend**

1. Ajuste Lineal: Se ajusta la señal a una línea recta:

<div style="text-align: center;">
    <h5>𝑓(𝑡)=𝑎⋅𝑡+𝑏</h5>
</div>

donde 𝑎 es la pendiente y 𝑏 es la intersección. Este ajuste se realiza utilizando mínimos cuadrados, donde se minimiza el error cuadrático:

<div style="text-align: center;">
    <h5>E = ∑(x(t_i) - (a · t_i + b))^2</h5>
</div>

​Para minimizar 𝐸, se derivan las ecuaciones normales:

<div style="text-align: center;">
    <h5>a = (N ∑(t_i x(t_i)) - ∑(t_i) ∑(x(t_i))) / (N ∑(t_i^2) - (∑(t_i))^2)</h5>
</div>

<div style="text-align: center;">
    <h5>b = (∑(x(t_i)) - a ∑(t_i)) / N</h5>
</div>

2. Detrending: Una vez obtenidos 𝑎 y 𝑏, se calcula la tendencia 𝑓(𝑡) y se sustrae de la señal original:

<div style="text-align: center;">
    <h5>𝑦(𝑡)=𝑥(𝑡)−𝑓(𝑡)=𝑥(𝑡)−(𝑎⋅𝑡+𝑏)</h5>
</div>

**•** Discutir matemáticamente, explicando con fórmulas, cómo funciona el filtro hampel **(5%)** 

El filtro Hampel es un método robusto para la detección y eliminación de outliers en señales y series temporales. Funciona basándose en la mediana y en el rango intercuartílico (IQR) para identificar valores atípicos.

**Concepto**

Dado un conjunto de datos 
𝑥=[𝑥1],[𝑥2],…,[𝑥𝑁],  el filtro Hampel busca reemplazar los valores atípicos (outliers) con un valor más representativo, generalmente la mediana de un vecindario.

**Definiciones**

1. Ventana de análisis: Para cada valor 𝑥𝑖, se considera una ventana centrada en 𝑥𝑖 , con un tamaño definido 𝑘. Esto significa que se toman los valores 𝑥𝑖−𝑘,𝑥𝑖−𝑘+1,…,𝑥𝑖+𝑘  para el análisis.

2. Mediana: La mediana 𝑚 de los valores en la ventana se calcula como:

<div style="text-align: center;">
    <h5>𝑚 = mediana(𝑥𝑖−𝑘,𝑥𝑖−𝑘+1,…,𝑥𝑖+𝑘)</h5>
</div>

3. Desviación absoluta: La desviación absoluta de cada elemento con respecto a la mediana se calcula como:

<div style="text-align: center;">
    <h5>dj​ = ∣xj − m∣ para j=i−k,…,i+k</h5>
</div>

4. Rango intercuartílico (IQR): El IQR se calcula como la diferencia entre el tercer cuartil 𝑄3 y el primer cuartil 𝑄1:

​<div style="text-align: center;">
    <h5>IQR=Q3 − Q1</h5>
</div>

5. Umbral: Se establece un umbral 𝑇 para identificar outliers. Generalmente se utiliza un multiplicador de la mediana de las desviaciones absolutas (MAD):

<div style="text-align: center;">
    <h5>T=α⋅MAD</h5>
</div>

donde 𝛼 es un parámetro que determina la sensibilidad del filtro.

**•**  Discutir matemáticamente y desde la bioingeniería qué mide la entropia de permutación **(5%)** 

La entropía de permutación es una medida utilizada para analizar la complejidad y el orden en series temporales, especialmente en el contexto de señales fisiológicas. Se basa en la idea de que la organización de los valores en una serie temporal puede proporcionar información sobre su dinámica subyacente.

**Concepto**
1. Permutaciones: Dada una serie temporal 𝑥=[𝑥1,𝑥2,…,𝑥𝑁], se consideran las 𝑚 dimensiones de las ventanas deslizantes de longitud 𝑚. Por ejemplo, para una serie 𝑥, se forman vectores de la siguiente manera:

<div style="text-align: center;">
    <h5>𝑋𝑚={(𝑥𝑖,𝑥𝑖+1,…,𝑥𝑖+𝑚−1)∣1≤𝑖≤𝑁−𝑚+1}</h5>
</div>

2. Orden de las permutaciones: Para cada vector de longitud 𝑚, se observa el orden de sus elementos. Se generan permutaciones de estos vectores que representan diferentes configuraciones de los elementos.

3. Cálculo de frecuencias: Se calcula la frecuencia de cada tipo de permutación en la serie temporal. Si hay 𝑘 tipos distintos de permutaciones, la frecuencia de la 𝑗-ésima permutación se denota como 𝑝𝑗.

4. Entropía de permutación: La entropía se define como:

<div style="text-align: center;">
    <h5>E = - ∑ 𝑝𝑗(log(𝑝𝑗))</h5>
</div>

donde 𝑝𝑗 es la probabilidad de la j-ésima permutación y 𝑘 es el número total de permutaciones observadas. Esta fórmula mide la incertidumbre o el desorden en la secuencia de permutaciones.

**Relevancia bioingeniería**

• Cuantificar la Variabilidad: Una alta entropía indica una mayor variabilidad en los patrones de movimiento, lo que podría sugerir una pérdida de control motor o una estrategia de marcha menos eficiente.

• Detectar Asincronías: Puede revelar asincronías en el movimiento de diferentes segmentos corporales, lo que es característico de algunas enfermedades neurológicas.

• Evaluar la Complejidad: Permite evaluar la complejidad de los patrones de marcha, lo que puede ser un indicador de la salud neurológica y muscular.
• Identificar Biomarcadores: La entropía de permutación puede servir como un biomarcador para diversas enfermedades, como el Parkinson, el Alzheimer y la enfermedad de Huntington.

**Limitaciones**

• Dependencia de los Parámetros: La elección del tamaño de la ventana y el número de permutaciones puede influir en los resultados.

• Sensibilidad al Ruido: El ruido en los datos puede afectar la estimación de la entropía.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

1. Crear una rutina que permita cargar un archivo .ts y extraiga las 4 señales de interés almacenelas en un ndarray.

```python
# Función para cargar y extraer las señales de interés
def load_and_extract_signals(file_path):
    names = [
        'Elapsed Time (sec)', 'Left Stride Interval (sec)', 'Right Stride Interval (sec)',
        'Left Swing Interval (sec)', 'Right Swing Interval (sec)', 'Left Swing Interval (% of stride)', 
        'Right Swing Interval (% of stride)', 'Left Stance Interval (sec)', 'Right Stance Interval (sec)',
        'Left Stance Interval (% of stride)', 'Right Stance Interval (% of stride)', 'Double Support Interval (sec)', 
        'Double Support Interval (% of stride)'
    ]
    
    data = pd.read_csv(file_path, delimiter='\t', names=names)
    
    columns_of_interest = [
        'Left Stride Interval (sec)', 'Right Stride Interval (sec)', 
        'Left Swing Interval (sec)', 'Right Swing Interval (sec)'
    ]
    
    signals = data[columns_of_interest].to_numpy()
    return signals
```

2. Para cada señal eliminar los errores en tendencia usando la rutina detrend.

```python
import scipy.signal as signal
def remove_trend(signals):
    detrended_signals = signal.detrend(signals, axis=0)
    return detrended_signals
```

3. Para cada señal resultante del paso 2 eliminar los datos atípicos usando filtro hampel.

```python
# Filtro de Hampel
def hampelfilter(signal: np.ndarray, w: int, sigma: int) -> np.ndarray:
    """
    Implementación del filtro de Hampel con ventana deslizante de tamaño w y umbral sigma.
    La entrada es una señal en forma de array de NumPy.
    """
    yk = np.copy(signal)  # Inicializa la señal filtrada (copia de la original).
    
    # Recorre la señal usando una ventana deslizante de tamaño w
    for i in range(w, len(signal) - w):
        ventana = signal[i - w:i + w + 1]
        ref = np.median(ventana)  # Calcula la mediana de la ventana.
        AbsDev = abs(ventana - ref)
        MAD = 1.4826 * np.median(AbsDev)  # Calcula el MAD para la ventana.
        
        TestVal = abs(signal[i] - ref)  # Diferencia del punto con la mediana.
        
        # Reemplaza el valor por la mediana si excede sigma * MAD
        if TestVal > sigma * MAD:
            yk[i] = ref

    return yk
```

4. Crear una función que reciba una señal de marcha y calcule el coeficiente de variación (CV) y la SD tal como se define en el artículo de referencia .

```python
def calculate_cv_sd(signal):
    mean_val = np.abs(np.mean(signal))
    sd_val = np.std(signal)
    cv_val = 100 * (sd_val / mean_val) 
    return cv_val, sd_val
```

5. Crear una función que reciba una señal de marcha y permita calcular la entropía de permutación

```python
import math
from collections import Counter
# Función para calcular la entropía de permutación
def calculate_permutation_entropy(time_series, D, tau):
    column_matrix = []
    for i in range(len(time_series) - (D - 1) * tau):
        column_vector = [time_series[i + j * tau] for j in range(D)]
        column_matrix.append(column_vector)

    permutations = []
    for column_vector in column_matrix:
        permutation = sorted(range(len(column_vector)), key=lambda k: column_vector[k])
        permutations.append(tuple(permutation))

    relative_frequencies = Counter(permutations)
    total_permutations = len(permutations)
    for permutation in relative_frequencies:
        relative_frequencies[permutation] /= total_permutations

    PE = -sum(f * math.log(f, D) for f in relative_frequencies.values())
    return PE, relative_frequencies
```

6. Crear una rutina que aplique sobre todos los archivos de la base de datos las rutina 1 y almacene los resultados en un dataframe donde se pueda discriminar nombre participante, condición (Parkinson, Huntington, ELA, Control) y los tres índices/características calculados por cada señal

```python
import re  # Importar la librería de expresiones regulares para extraer los números

# Función para ordenar archivos numéricamente
def numerical_sort(value):
    """ Extrae el número del nombre del archivo para ordenarlo numéricamente. """
    numbers = re.findall(r'\d+', value)  # Busca todos los números en el nombre del archivo
    return int(numbers[0]) if numbers else -1  # Convierte el número a entero y lo devuelve

# Función para procesar todos los sujetos en una carpeta
def process_all_subjects(folder_path, title):
    results = []

    # Obtener lista de archivos ordenados numéricamente
    files = sorted([f for f in os.listdir(folder_path) if f.endswith('.ts')], key=numerical_sort)

    for file_name in files:
        file_path = os.path.join(folder_path, file_name)
        
        # Cargar y extraer las señales
        signals = load_and_extract_signals(file_path)
        
        # Remover tendencia
        detrended_signals = remove_trend(signals)
        
        # Eliminar valores atípicos
        filtered_signals = np.zeros_like(detrended_signals)
        for i in range(detrended_signals.shape[1]):
            filtered_signals[:, i] = hampelfilter(detrended_signals[:, i], w=5, sigma=3)
        
        # Calcular características para cada señal
        cv_left_stride, sd_left_stride = calculate_cv_sd(filtered_signals[:, 0])
        ent_left_stride, _ = calculate_permutation_entropy(filtered_signals[:, 0], D=3, tau=1)
        
        cv_right_stride, sd_right_stride = calculate_cv_sd(filtered_signals[:, 1])
        ent_right_stride, _ = calculate_permutation_entropy(filtered_signals[:, 1], D=3, tau=1)
        
        cv_left_swing, sd_left_swing = calculate_cv_sd(filtered_signals[:, 2])
        ent_left_swing, _ = calculate_permutation_entropy(filtered_signals[:, 2], D=3, tau=1)
        
        cv_right_swing, sd_right_swing = calculate_cv_sd(filtered_signals[:, 3])
        ent_right_swing, _ = calculate_permutation_entropy(filtered_signals[:, 3], D=3, tau=1)
        
        # Almacenar los resultados en un diccionario con el formato solicitado
        results.append({
            'Subject': file_name.split('.')[0],  # Nombre del archivo sin extensión
            'Condition': title,
            'Entropy_Left Stride Interval': ent_left_stride,
            'CV_Left Stride Interval': cv_left_stride,
            'SD_Left Stride Interval': sd_left_stride,
            'Entropy_Right Stride Interval': ent_right_stride,
            'CV_Right Stride Interval': cv_right_stride,
            'SD_Right Stride Interval': sd_right_stride,
            'Entropy_Left Swing Interval': ent_left_swing,
            'CV_Left Swing Interval': cv_left_swing,
            'SD_Left Swing Interval': sd_left_swing,
            'Entropy_Right Swing Interval': ent_right_swing,
            'CV_Right Swing Interval': cv_right_swing,
            'SD_Right Swing Interval': sd_right_swing
        })
    
    # Convertir los resultados en un DataFrame
    results_df = pd.DataFrame(results)
    return results_df
```

```python
import seaborn as sns
# Función para crear un subplot con 3 columnas y 2 filas, con estilos y colores mejorados
def boxplot_subplots(df, measures_left, measures_right, titles_left, titles_right):
    sns.set(style="whitegrid")  # Estilo de fondo blanco con grid
    palette = "coolwarm"  # Paleta de colores 

    fig, axes = plt.subplots(2, 3, figsize=(18, 12))  # 3 columnas, 2 filas
    
    # Fila 1: Métricas de la pierna izquierda
    for i, measure in enumerate(measures_left):
        sns.boxplot(x='Condition', y=measure, hue='Condition', data=df, ax=axes[0, i], palette=palette, width=0.5, linewidth=1.5, dodge=False,showfliers=False)
        axes[0, i].set_title(f'{titles_left[i]} by Condition', fontsize=14, fontweight='bold')
        axes[0, i].set_xlabel('Condition', fontsize=12)
        axes[0, i].set_ylabel(measure, fontsize=12)
        axes[0, i].tick_params(axis='x', rotation=45)  # Rotar etiquetas del eje x
    
    # Fila 2: Métricas de la pierna derecha
    for i, measure in enumerate(measures_right):
        sns.boxplot(x='Condition', y=measure, hue='Condition', data=df, ax=axes[1, i], palette=palette, width=0.5, linewidth=1.5, dodge=False,showfliers=False)        
        axes[1, i].set_title(f'{titles_right[i]} by Condition', fontsize=14, fontweight='bold')
        axes[1, i].set_xlabel('Condition', fontsize=12)
        axes[1, i].set_ylabel(measure, fontsize=12)
        axes[1, i].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

# Función para filtrar valores anómalos y NaN
def filter_data(series, lower_percentile=1, upper_percentile=99):
    lower_bound = np.percentile(series.dropna(), lower_percentile)
    upper_bound = np.percentile(series.dropna(), upper_percentile)
    return series[(series >= lower_bound) & (series <= upper_bound)]

# Función para crear subplots con métricas en 3 columnas y 2 filas
def histplot_subplots(df, measures_left, measures_right, titles_left, titles_right):
    sns.set(style="whitegrid")

    # Obtener los colores únicos de la columna 'Condition'
    unique_conditions = df['Condition'].unique()
    custom_palette = sns.color_palette("husl", len(unique_conditions))  # Paleta ajustada

    # Crear subplots de 2 filas y 3 columnas
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))

    # Iterar sobre las métricas del lado izquierdo
    for i, measure in enumerate(measures_left):
        # Filtrar datos
        x_left = filter_data(df[measure].dropna())
        
        # Histograma con KDE
        sns.histplot(x_left, kde=True, ax=axes[0, i], color=custom_palette[i % len(custom_palette)])
        
        # Línea vertical en la media
        axes[0, i].axvline(np.mean(x_left), color='red', linestyle='--')  
        
        # Título de cada gráfico
        axes[0, i].set_title(titles_left[i], fontsize=14, fontweight='bold')

    # Iterar sobre las métricas del lado derecho
    for i, measure in enumerate(measures_right):
        # Filtrar datos
        x_right = filter_data(df[measure].dropna())
        
        # Histograma con KDE
        sns.histplot(x_right, kde=True, ax=axes[1, i], color=custom_palette[i % len(custom_palette)])
        
        # Línea vertical en la media
        axes[1, i].axvline(np.mean(x_right), color='red', linestyle='--')  
        
        # Título de cada gráfico
        axes[1, i].set_title(titles_right[i], fontsize=14, fontweight='bold')

    plt.tight_layout()
    plt.show()
```

<h8><strong>b.</strong></h8> <h8>Planteamiento de las hipótesis nulas y alternativas(5%), selección del tipo de prueba (paramétrica o no paramétrica)(10%) y discusión de los resultados.(10%)</h8>

```python
from scipy import stats

# Define las métricas que deseas analizar
metrics = ['Entropy_Left Stride Interval', 'CV_Left Stride Interval', 'SD_Left Stride Interval',
           'Entropy_Right Stride Interval', 'CV_Right Stride Interval', 'SD_Right Stride Interval',
           'Entropy_Left Swing Interval', 'CV_Left Swing Interval', 'SD_Left Swing Interval',
           'Entropy_Right Swing Interval', 'CV_Right Swing Interval', 'SD_Right Swing Interval']

# Inicializa una lista para almacenar los resultados
results = []

# Realiza pruebas para cada métrica
for metric in metrics:
    # Filtrar los datos
    data = final_df[metric].dropna()
    
    # Prueba de normalidad (Shapiro-Wilk)
    stat_normality, p_value_normality = stats.shapiro(data)
    normality_result = 'Normal' if p_value_normality > 0.05 else 'No normal'

    # Prueba de homocedasticidad (Levene)
    # Aquí asumo que tienes varias condiciones para comparar. Ajusta según tus datos.
    conditions_data = [final_df[final_df['Condition'] == cond][metric].dropna() for cond in final_df['Condition'].unique()]
    stat_homoscedasticity, p_value_homoscedasticity = stats.levene(*conditions_data)
    homoscedasticity_result = 'Homocedástico' if p_value_homoscedasticity > 0.05 else 'Heterocedástico'

    # Agrega los resultados a la lista
    results.append({
        'Metric': metric,
        'Normality (p-value)': p_value_normality,
        'Normality Result': normality_result,
        'Homoscedasticity (p-value)': p_value_homoscedasticity,
        'Homoscedasticity Result': homoscedasticity_result
    })

# Crea un DataFrame con los resultados
results_df = pd.DataFrame(results)

# Muestra el DataFrame
results_df
```

```python
from scipy.stats import mannwhitneyu
def mann_whitney_comparison(df, measure, alpha=0.05):
    results = {}
    control_group = df[df['Condition'] == 'Control'][measure]

    # Comparar con Parkinson
    park_group = df[df['Condition'] == 'Parkinson'][measure]
    stat, p_value = mannwhitneyu(control_group, park_group)
    results['Parkinson'] = {
        'p-value': p_value,
        'Difference': 'Diferencia significativa' if p_value < alpha else 'Sin diferencia significativa'
    }

    # Comparar con ELA
    ela_group = df[df['Condition'] == 'ELA'][measure]
    stat, p_value = mannwhitneyu(control_group, ela_group)
    results['ELA'] = {
        'p-value': p_value,
        'Difference': 'Diferencia significativa' if p_value < alpha else 'Sin diferencia significativa'
    }

    # Comparar con Huntington
    hunt_group = df[df['Condition'] == 'Huntington'][measure]
    stat, p_value = mannwhitneyu(control_group, hunt_group)
    results['Huntington'] = {
        'p-value': p_value,
        'Difference': 'Diferencia significativa' if p_value < alpha else 'Sin diferencia significativa'
    }

    return results

# Lista de métricas
measures = ['Entropy_Left Stride Interval', 'CV_Left Stride Interval', 'SD_Left Stride Interval',
           'Entropy_Right Stride Interval', 'CV_Right Stride Interval', 'SD_Right Stride Interval',
           'Entropy_Left Swing Interval', 'CV_Left Swing Interval', 'SD_Left Swing Interval',
           'Entropy_Right Swing Interval', 'CV_Right Swing Interval', 'SD_Right Swing Interval']

# Crear un diccionario para almacenar resultados
results_dict = {measure: {} for measure in measures}

# Llenar el diccionario con resultados
for measure in measures:
    comparison_results = mann_whitney_comparison(final_df, measure)
    for condition, result in comparison_results.items():
        results_dict[measure][condition] = result

# Crear DataFrame final
results_df = pd.DataFrame.from_dict({(i, j): results_dict[i][j] 
                                      for i in results_dict.keys() 
                                      for j in results_dict[i].keys()},
                                     orient='index')

# Reorganizar el DataFrame para que solo muestre p-value y el resultado
results_df = results_df[['p-value', 'Difference']]

# Mostrar DataFrame final
results_df

```

<h8><strong>d.</strong></h8> <h8>Proponer cuales de las medidas usadas podrían servir para discriminar que poblaciones clínicas de los controles.(10%)</h8>

Las medidas que permiten discriminar poblaciones clínicas de poblaciones de control, según los resultados de esta investigación y el artículo base, incluyen:

- Tiempo de Zancada Promedio: Un tiempo de zancada significativamente más largo en pacientes con ELA en comparación con los controles indica una marcha alterada.

- Variabilidad en el Tiempo de Zancada: La mayor variabilidad (fluctuaciones de zancada a zancada) en pacientes con ELA es notablemente mayor que en los controles, lo que sugiere dificultades en el control motor.

- Duración del Ciclo de Marcha: Un ciclo de marcha prolongado en los pacientes con ELA en comparación con los controles también es un indicador de alteraciones motoras.

- Entropía de Permutación: Esta medida evalúa la complejidad y la irregularidad de la señal de marcha. Un aumento en la entropía de permutación en pacientes con enfermedades neurodegenerativas sugiere una marcha menos predecible y más caótica, lo que puede indicar un deterioro en el control motor.

- Coeficiente de Variación (CV): El CV es la relación entre la desviación estándar y la media, proporcionando una medida de la variabilidad relativa. Un CV elevado en los intervalos de zancada en pacientes con enfermedades neurodegeneativas en comparación con los controles indica una mayor inestabilidad y variabilidad en la marcha, lo que puede ser indicativo de problemas en la coordinación y el control motor.

- SD (Desviación Estándar) de los Intervalos de Zancada: Un aumento en la desviación estándar de los intervalos de zancada puede ser un indicador de inestabilidad y dificultades en la marcha.

- Índice de No Estacionariedad (NSI): Aunque en este análisis no se mencionó, este índice se utiliza para evaluar la estabilidad de los patrones de marcha, siendo útil para diferenciar entre poblaciones.

- Pruebas de Normalidad y Homocedasticidad: Estas pruebas ayudan a determinar la adecuación de los métodos estadísticos y a asegurar que las diferencias observadas sean significativas y confiables.
